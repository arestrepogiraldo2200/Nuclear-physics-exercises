# Magnetic dipole moments (Long Lived)

```python
M_LL_0 = pd.read_csv("./Stone.2019.indc-nds-0794-pages.csv") 
M_LL_0 = M_LL_0.dropna()
#M_LL_0
```

```python
# Data organized
M_LL = pd.DataFrame(columns=["Z", "Sym", "A"] + M_LL_0.columns[0].split()[1:])
misbehavedcount = 0
totalrows = 0

for i in range(len(M_LL_0)):
     
    entry = str(M_LL_0.iloc[i,0]).split()
    #ZSymA = [entry[0], entry[1], entry[2]]
    
    # Page number in pdfs discarded  <---------------------------- here are some of the discarded rows
    if len(entry) <= 3:
        totalrows += 1
        continue
        
    # Actual entries
    else:
        
        # Check format of good entry
        try:
            
            # First entry must be integer
            Z = int(entry[0])

            # Second entry must be the element symbol
            if entry[1] in elements:

                if entry[4] == "stable":
                    totalrows += 1
                    M_LL = M_LL.append(pd.Series([entry[0], entry[1], entry[2], entry[3], entry[4], entry[5], entry[6], entry[7], entry[8], "".join(entry[9:])], index=M_LL.columns), ignore_index=True)
                else:
                    totalrows += 1
                    M_LL = M_LL.append(pd.Series([entry[0], entry[1], entry[2], entry[3], "".join(entry[4:6]), entry[6], entry[7], entry[8], entry[9], "".join(entry[10:])], index=M_LL.columns), ignore_index=True)

            # Misbehaving lines
            else:
                #print(entry)
                totalrows += 1
                misbehavedcount += 1
                    
        except:
            # Misbehaving lines
            #print(entry) 
            totalrows += 1
            misbehavedcount += 1
            
print("Number of misbehaved entries: ", misbehavedcount)
print("Total number of entries analyzed: ", totalrows)
print("Total number of entries uncleaned: ", len(M_LL_0))
print("Total number of entries inserted: ", len(M_LL))
```

```python
print("==============================================================================")
print("==============================================================================")
print("Long lived magnetic dipole moments loaded as M_LL")
print("Extract data with the function extract_mu(D_F), deliver the dataframe D_F as argument. The return is a list [Z, A, mu, un].")
print("==============================================================================")
print("==============================================================================")
```

```python
pattern1 = r"\([+-]\)\d+\.\d+\(\d+\)"                 # Pattern (sign)-measure-(error)
sign1    = r"\([+-]\)"
value1   = r"\)\d+\.\d+\("
error1   = r"\(\d+\)"

pattern2     = r"[+-]{1}\d+\.\d+\(\d+\)"              # Pattern sign-measure-(error)            
sign_value2  = r"[+-]{1}\d+\.\d+\("
error2       = r"\(\d+\)"

pattern3 = r"\d+\.\d+\(\d+\)"                         # Pattern measure-(error)
value3   = r"\d+\.\d+\("
error3   = r"\(\d+\)"

pattern4  = r"\d+\.\d+\([+-]{1}\d+\,[+-]\d+\)"        # Pattern measure-(+error_u, -error_d)
value4    = r"\d+\.\d+\("
error4    = r"\([+-]{1}\d+\,[+-]{1}\d+\)"

pattern5  = r"\([+-]\)\d+\(\d+\)"                     # Pattern (sign)-measure integer-(error)
sign5     = r"\([+-]\)"
value5    = r"\)\d+\("
error5    = r"\(\d+\)"

pattern6  = r"[+-]\d+\(\d+\)"                         # Pattern sign-measure integer-(error)
sign6     = r"[+-]"
value6    = r"[+-]\d+\("
error6    = r"\(\d+\)"

pattern7  = r"[+-]{1}\d+\.\d+\([+-]{1}\d+\,[+-]\d+\)" # Pattern sign-measure-(+error_u, -error_d)
sign7     = r"[+-]{1}\d+\."
value7    = r"\d+\.\d+\("
error7    = r"\([+-]{1}\d+\,[+-]{1}\d+\)"

pattern8  = r"\d+\(\d+\)"                             # Pattern measure integer-(integer error)
value8    = r"\d+\("
error8    = r"\(\d+\)"

# =========================================================================================
# M_LL: Pandas dataframe from database

def extract_mu(M_LL):
    
    Z  = []
    A  = []
    mu = []
    un = []
    E  = []
    jp = []
    
    for i in range(len(M_LL)):

        VV = M_LL.iloc[i]["m(nm)"]

        if re.match(pattern1, VV):

            s = re.findall(sign1, VV)[0][1:-1]
            v = re.findall(value1, VV)[0][1:-1]
            e = re.findall(error1, VV)[0][1:-1]
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])

        elif re.match(pattern2, VV):

            s = re.findall(sign_value2, VV)[0][0]
            v = re.findall(sign_value2, VV)[0][1:-1]
            e = re.findall(error2, VV)[0][1:-1]

            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern3, VV):

            s = "+"
            v = re.findall(value3, VV)[0][0:-1]
            e = re.findall(error3, VV)[0][1:-1]
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern4, VV):

            s = "+"
            v = re.findall(value4, VV)[0][0:-1]
            e = re.findall(error4, VV)[0][1:-1].split(",")
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v))
            order = len(v.split(".")[1])
            un.append([int(e[0])/10**order, int(e[1])/10**order])        # This uncertainty is a list
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern5, VV):

            s = re.findall(sign5, VV)[0][1:-1]
            v = re.findall(value5, VV)[0][1:-1]
            e = re.findall(error5, VV)[0][1:-1]
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v)) 
            order = 0
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern6, VV):

            s = re.findall(sign6, VV)[0]
            v = re.findall(value6, VV)[0][1:-1]
            e = re.findall(error6, VV)[0][1:-1]
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v)) 
            order = 0
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern7, VV):

            s = re.findall(sign7, VV)[0][0]
            v = re.findall(value7, VV)[0][0:-1]
            e = re.findall(error7, VV)[0][1:-1].split(",")
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v))
            order = len(v.split(".")[1])
            un.append([int(e[0])/10**order, int(e[1])/10**order])        # This uncertainty is a list 2
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        elif re.match(pattern8, VV):

            s = "+"
            v = re.findall(value8, VV)[0][0:-1]
            e = re.findall(error8, VV)[0][1:-1]
            
            Z.append(int(M_LL.iloc[i]["Z"]))
            A.append(int(M_LL.iloc[i]["A"]))
            mu.append(float(s+v))
            order = 0
            un.append(float(e)/10**order)
            try:
                E.append(float(M_LL.iloc[i]["Ex"]))
            except:
                E.append(np.nan)
            jp.append(M_LL.iloc[i]["Jπ"])
            
        else:
            print("Data with none of the available formats:")
            display(M_LL.iloc[i])
            
    return [Z, A, mu, un, E, jp]
```

```python
qpattern1 = r"\([+-]\)\d+\.\d+\(\d+\)"                # pattern (sign)-measure-(error)
qsign1    = r"\([+-]\)"
qvalue1   = r"\)\d+\.\d+\("
qerror1   = r"\(\d+\)"

qpattern2     = r"[+-]{1}\d+\.\d+\(\d+\)"             # pattern sign-measure-(qerror)            
qsign_qvalue2  = r"[+-]{1}\d+\.\d+\("
qerror2       = r"\(\d+\)"

qpattern3 = r"\d+\.\d+\(\d+\)"                        # pattern measure-(error)
qvalue3   = r"\d+\.\d+\("
qerror3   = r"\(\d+\)"

qpattern4  = r"\d+\.\d+\([+-]{1}\d+\,[+-]\d+\)"        # pattern measure-(+error_u, -error_d)
qvalue4    = r"\d+\.\d+\("
qerror4    = r"\([+-]{1}\d+\,[+-]{1}\d+\)"

qpattern5  = r"\([+-]\)\d+\.\d+"                       # pattern (sign)-measure      NO error
qsign5     = r"\([+-]\)"
qvalue5    = r"\)\d+\.\d+"

qpattern6  = r"\d+\.\d+"                               # pattern (sign)-measure     NO error

qpattern7  = r"[+-]\d+\(\d+\)"                         # pattern sign- integer measure 
qsign7     = r"[+-]"
qvalue7    = "[+-]\d+\("
qerror7   = r"\(\d+\)"

# =========================================================================================
# Q: Pandas dataframe from database

def extract_q(Q):
    
    Z  = []
    A  = []
    q  = []
    un = []
    E  = []
    jp = []
    
    for i in range(len(Q)):

        VV = Q.iloc[i]["Q(b)"]

        if re.match(qpattern1, VV):

            s = re.findall(qsign1, VV)[0][1:-1]
            v = re.findall(qvalue1, VV)[0][1:-1]
            e = re.findall(qerror1, VV)[0][1:-1]
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern2, VV):

            s = re.findall(qsign_qvalue2, VV)[0][0]
            v = re.findall(qsign_qvalue2, VV)[0][1:-1]
            e = re.findall(qerror2, VV)[0][1:-1]

            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern3, VV):

            s = "+"
            v = re.findall(qvalue3, VV)[0][0:-1]
            e = re.findall(qerror3, VV)[0][1:-1]
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v)) 
            order = len(v.split(".")[1])
            un.append(float(e)/10**order)
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern4, VV):

            s = "+"
            v = re.findall(qvalue4, VV)[0][0:-1]
            e = re.findall(qerror4, VV)[0][1:-1].split(",")
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v))
            order = len(v.split(".")[1])
            un.append([int(e[0])/10**order, int(e[1])/10**order])        # This uncertainty is a list
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern5, VV):

            s = re.findall(qsign5, VV)[0][1:-1]
            v = re.findall(qvalue5, VV)[0][1:]
            e = 0
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v))
            un.append(0)                                                 # No uncertainty reported
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern6, VV):

            s = "+"
            v = re.findall(qpattern6, VV)[0][0:]
            e = 0
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v))
            un.append(0)
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        elif re.match(qpattern7, VV):

            s = re.findall(qsign7, VV)[0][0]
            v = re.findall(qvalue7, VV)[0][1:-1]
            e = re.findall(qerror7, VV)[0][1:-1]
            
            Z.append(int(Q.iloc[i]["Z"]))
            A.append(int(Q.iloc[i]["A"]))
            q.append(float(s+v)) 
            order = 0
            un.append(float(e)/10**order)
            try:
                E.append(float(Q.iloc[i]["E(level)"]))
            except:
                E.append(np.nan)
            jp.append(Q.iloc[i]["I"])
                
        else:
            print("None of the available formats:")
            display(Q.iloc[i])
            
    return [Z, A, q, un, E, jp]
```
